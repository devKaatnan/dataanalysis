```python
#insert new empty column to categorize  at index 4 thats after the BS column
df.insert(4,'BS Cat', '')
```

```python
#inserts value based on the BS column
for i in range(len(df)):
    if df.at[i,'BS'] < 7:
        df.at[i,'BS Cat'] = 'Low'
    elif df.at[i,'BS'] >= 7 and df['BS'][i] <= 10:
        df.at[i,'BS Cat'] = 'Normal'
    else:
        df.at[i,'BS Cat'] = 'High'
```
